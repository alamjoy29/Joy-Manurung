```python
#pengelompokan berdasarkan kondisi cuaca seperti suhu, kelembaban, dan kecepatan angin, dan kemudian menghitung statistik seperti jumlah peminjaman sepeda (cnt) untuk setiap kelompok.

import pandas as pd

# Dataframe yang berisi data penggunaan sepeda sewa
day_df = pd.read_csv("C:/Users/ASUS/OneDrive/Documents/MBKM/BANGKIT/submission_joy/data/day.csv")  

# Membuat fungsi untuk menentukan kategori kecepatan angin
def wind_speed_category(speed):
    if speed <= 0.2:
        return 'Calm'
    elif speed <= 1.5:
        return 'Light air'
    elif speed <= 3.3:
        return 'Light breeze'
    elif speed <= 5.4:
        return 'Gentle breeze'
    elif speed <= 7.9:
        return 'Moderate breeze'
    elif speed <= 10.7:
        return 'Fresh breeze'
    elif speed <= 13.8:
        return 'Strong breeze'
    elif speed <= 17.1:
        return 'Moderate gale'
    elif speed <= 20.7:
        return 'Fresh gale'
    elif speed <= 24.4:
        return 'Strong gale'
    elif speed <= 28.4:
        return 'Whole gale'
    elif speed <= 32.6:
        return 'Storm'
    else:
        return 'Hurricane'

# Menambahkan kolom baru untuk kategori kecepatan angin
day_df['wind_speed_category'] = day_df['windspeed'].apply(wind_speed_category)

# Mengelompokkan data berdasarkan suhu, kelembaban, dan kategori kecepatan angin, 
# kemudian menghitung jumlah peminjaman sepeda untuk setiap kelompok
weather_effect = day_df.groupby(by=['temp', 'hum', 'wind_speed_category']).agg({
    'cnt': 'sum'
}).reset_index()

# Menampilkan tabel hasil analisis explanatory
print(weather_effect)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana distribusi penggunaan sepeda sewa tiap seasonnya?

```python
# Menghitung total peminjaman sepeda sepanjang season

day_df = pd.read_csv("C:/Users/ASUS/OneDrive/Documents/MBKM/BANGKIT/submission_joy/data/day.csv")  
total_rentals_season1 = day_df[day_df['season'] == 1]['cnt'].sum()
total_rentals_season2 = day_df[day_df['season'] == 2]['cnt'].sum()
total_rentals_season3 = day_df[day_df['season'] == 3]['cnt'].sum()
total_rentals_season4 = day_df[day_df['season'] == 4]['cnt'].sum()


print("Total Peminjaman Sepeda season pertama:", total_rentals_season1)
print("Total Peminjaman Sepeda season kedua:", total_rentals_season2)
print("Total Peminjaman Sepeda season ketiga:", total_rentals_season3)
print("Total Peminjaman Sepeda season keempat:", total_rentals_season4)

# Menghitung jumlah peningkatan
peningkatan = total_rentals_season4 - total_rentals_season1
print("Jumlah Peningkatan Peminjaman Sepeda:", peningkatan)

# Menghitung persentase peningkatan
persentase_peningkatan = (peningkatan / total_rentals_season1) * 100
print("Persentase Peningkatan Peminjaman Sepeda:", persentase_peningkatan, "%")

season_labels = ['season 1', 'season 2', 'season 3', 'season 4']
total_rentals_seasons = [total_rentals_season1, total_rentals_season2, total_rentals_season3, total_rentals_season4]

plt.figure(figsize=(10, 5))
plt.bar(season_labels, total_rentals_seasons, color='skyblue')
plt.title('Total Peminjaman Sepeda per Season')
plt.xlabel('Season')
plt.ylabel('Total Peminjaman Sepeda')
plt.grid(True)
plt.show()
```

### Pertanyaan 2 Apakah kondisi cuaca seperti suhu, kelembaban, dan kecepatan angin memengaruhi jumlah peminjaman sepeda? bagaimana persebaran penguna sewa sepeda dalam tiap kondisi? 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
day_df = pd.read_csv("C:/Users/ASUS/OneDrive/Documents/MBKM/BANGKIT/submission_joy/data/day.csv")

# Scatter plot untuk analisis pengaruh suhu dan kelembaban terhadap jumlah peminjaman sepeda
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='hum', size='cnt', data=day_df, alpha=0.7)
plt.title('Pengaruh Suhu dan Kelembaban terhadap Jumlah Peminjaman Sepeda')
plt.xlabel('Suhu (Celsius)')
plt.ylabel('Kelembaban')
plt.grid(True)
plt.show()
```
